importing dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection

```python
#loading the datasheet to pandas dataframe
heart_data=pd.read_csv("/content/heart_disease_data.csv")
#showing the first five rows of datasheet
heart_data.head()
```

```python
#showing the last five rows of datasheeet
heart_data.tail()
```

```python
#no of rows and column of datasheest
heart_data.shape
```

```python
#getting some info about the data
heart_data.info()
```

```python
# checking any null value available in the column
heart_data.isnull().sum()
```

```python
#statistical value of the data
heart_data.describe()
```

```python
#checking the distribution of the target value
heart_data['target'].value_counts()
```

1----> herat disease
0----> healty heart

```python
#splitting the data into features and target
x=heart_data.drop(columns='target',axis=1)
y=heart_data['target']
print(x)
print(y)
```

```python
#splitting the data into training and testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
print(x_train.shape,x.shape,x_test.shape)
```

Model training


```python
model = LogisticRegression(max_iter=10000)
#traing the model with training data
model.fit(x_train,y_train)

```

Accuracy score of training data



```python
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
print('Accuracy on training data :',training_data_accuracy)
```

accuracy on the test data


```python
x_test_prediction=model.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
print('Accuracy on testing data :',testing_data_accuracy)
```

building a pridictive system

```python
input_data=(67,1,0,160,286,0,0,108,1,1.5,1,3,2)
#changing input data as numpy array
input_data_as_numpyarray=np.asarray(input_data)
#reshaping the input data
input_data_reshaped=input_data_as_numpyarray.reshape(1,-1)
prediction=model.predict(input_data_reshaped)
print(prediction)
if prediction[0]==1:
  print("the pesong has a heart defect")
else:
  print("the person is normal")
```
